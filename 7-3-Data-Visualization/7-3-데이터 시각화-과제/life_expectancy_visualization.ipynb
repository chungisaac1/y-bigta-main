```python
import pandas as pd 

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = 'Life Expectancy.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.describe()
```

```python
# 1. 검증 하고자 하는 가설 혹은 질문 > 국가의 개발 상태에 따라 기대수명, 성인 사망률, 유아 사망률 차이가 있을 것이고 차이의 원인은 개발도상국의 Alcohol 과다 섭취일 것이다
# 2. 독립변수 : 국가의 개발 상태 
# 2. 종속변수 : 기대수명, 성인 사망률, 유아 사망률 ,Alchohol
# 2. 데이터 특성 : status 열은 각 국가가 개발도상국인지 개발된 국가인지를 나타냄 Developing, Developed 
```

```python
# 시각화 1 : 국가의 개발 상태에 따른 기대수명 분포 
selected_columns = ['Status', 'Life expectancy ', 'Adult Mortality', 'infant deaths']
data = df[selected_columns]

# 결측값 처리
data.dropna(subset=['Life expectancy ', 'Adult Mortality', 'infant deaths'], inplace=True)

# 1. 국가의 개발 상태에 따른 기대수명 분포
plt.figure(figsize=(12, 6))
sns.boxplot(x='Status', y='Life expectancy ', data=data)
plt.title('Life Expectancy by Development Status')
plt.xlabel('Development Status')
plt.ylabel('Life Expectancy (years)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Status', y='Adult Mortality', data=data)
plt.title('Adult Mortality by Development Status')
plt.xlabel('Development Status')
plt.ylabel('Adult Mortality (per 1000)')
plt.grid(True)
plt.show()
```

```python
# 3. 국가의 개발 상태에 따른 유아 사망률 분포
plt.figure(figsize=(12, 6))
sns.boxplot(x='Status', y='infant deaths', data=data)
plt.title('Infant Mortality by Development Status')
plt.xlabel('Development Status')
plt.ylabel('Infant Mortality (per 1000)')
plt.grid(True)
plt.show()
```

```python
# 서브플롯 설정
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 1. 국가의 개발 상태에 따른 기대수명 분포
sns.boxplot(x='Status', y='Life expectancy ', data=data, ax=axes[0])
axes[0].set_title('Life Expectancy by Development Status')
axes[0].set_xlabel('Development Status')
axes[0].set_ylabel('Life Expectancy (years)')

# 2. 국가의 개발 상태에 따른 성인 사망률 분포
sns.boxplot(x='Status', y='Adult Mortality', data=data, ax=axes[1])
axes[1].set_title('Adult Mortality by Development Status')
axes[1].set_xlabel('Development Status')
axes[1].set_ylabel('Adult Mortality (per 1000)')

# 3. 국가의 개발 상태에 따른 유아 사망률 분포
sns.boxplot(x='Status', y='infant deaths', data=data, ax=axes[2])
axes[2].set_title('Infant Mortality by Development Status')
axes[2].set_xlabel('Development Status')
axes[2].set_ylabel('Infant Mortality (per 1000)')


# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python


# 필요한 열만 선택
selected_columns = ['Status', 'Life_expectancy_', 'Adult_Mortality', 'infant_deaths', 'Alcohol']
data = df[selected_columns]

# 결측값 처리
data.dropna(subset=['Life_expectancy_', 'Adult_Mortality', 'infant_deaths', 'Alcohol'], inplace=True)

# 서브플롯 설정
fig, axes = plt.subplots(1, 4, figsize=(24, 6))

# 1. 국가의 개발 상태에 따른 기대수명 분포
sns.boxplot(x='Status', y='Life_expectancy_', data=data, ax=axes[0])
axes[0].set_title('Life Expectancy by Development Status')
axes[0].set_xlabel('Development Status')
axes[0].set_ylabel('Life Expectancy (years)')

# 2. 국가의 개발 상태에 따른 성인 사망률 분포
sns.boxplot(x='Status', y='Adult_Mortality', data=data, ax=axes[1])
axes[1].set_title('Adult Mortality by Development Status')
axes[1].set_xlabel('Development Status')
axes[1].set_ylabel('Adult Mortality (per 1000)')

# 3. 국가의 개발 상태에 따른 유아 사망률 분포
sns.boxplot(x='Status', y='infant_deaths', data=data, ax=axes[2])
axes[2].set_title('Infant Mortality by Development Status')
axes[2].set_xlabel('Development Status')
axes[2].set_ylabel('Infant Mortality (per 1000)')

# 4. 국가의 개발 상태에 따른 알코올 소비량 분포
sns.boxplot(x='Status', y='Alcohol', data=data, ax=axes[3])
axes[3].set_title('Alcohol Consumption by Development Status')
axes[3].set_xlabel('Development Status')
axes[3].set_ylabel('Alcohol Consumption (liters per capita)')

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 3번 가설/질문에 대한 결론 
# 시각화 결과 
# 개발된 국가(Developed)는 기대수명이 상대적으로 높았고 중앙값은 약 80년이다 
# 개발된 국가 는 성인 사망률이 더 낮았다 
# 개발된 국가의 유아 사망률은 매우 낮았으며 대부분 1000명당 50명 이하였다 . 
# 개발된 국가의 알코올 소비량이 더 높은 경향을 보이며 중앙값이 약 10리터이다 

# 질문에 대한 결론 : 개발된 국가에서 기대수명이 높고 성인 및 유아 사망률이 낮았고 오히려 알코올 소비량이 더 높았다. 
```
